```python
directory = '/content/gdrive/MyDrive/DeepLearning/Dataset2B'

className = sorted(os.listdir(directory))
num_classes = len(className)

img_size = (64, 64, 3)

print('classes: ', className)
```

```python
#save Data Class
labels = []
images = []

print('images:\n')
for pathClass in className:
    print(pathClass, end=' -> ')
    for img in os.listdir(directory+ '/' +  pathClass):
        label = np.zeros(num_classes)
        label[className.index(pathClass)] = 1
        labels.append(label)
        
        image = np.asarray(cv2.resize(cv2.imread(directory + '/' + pathClass + '/' + img, cv2.IMREAD_COLOR), img_size[0:2]))
        images.append(image)
    print('done')
```

```python
labels = np.asarray(labels)
images = np.asarray(images)

print(f'\n\nlabels shape: {labels.shape}')
print(f'images shape: {images.shape}')
```

b. [LO 3, 5 poin] Pisahkan dataset menjadi 80% training set, 10% validation set dan 10% test set

```python
x_train, x_valid, y_train, y_valid = train_test_split(images, labels, test_size = 0.1, random_state = 42)
x_train, x_test, y_train, y_test = train_test_split(x_train, y_train, test_size = 0.1, random_state = 42)
```

```python
print(f'train images shape: {x_train.shape}\ntrain labels shape: {y_train.shape}\n\nvalidation images shape: {x_valid.shape}\nvalidation labels shape: {y_valid.shape}\n')
print(f'test images shape: {x_test.shape}\ntrain labels shape: {y_test.shape}\n\nvalidation images shape: {x_valid.shape}\nvalidation labels shape: {y_valid.shape}\n')
```

```python
# data augmentation
trainGenerator = tf.keras.preprocessing.image.ImageDataGenerator(shear_range=0.2,
                                                                         zoom_range=0.2,
                                                                         vertical_flip=True)
trainGenerator = trainGenerator.flow(x_train, y=y_train)

valGenerator = tf.keras.preprocessing.image.ImageDataGenerator(vertical_flip=True)
valGenerator = valGenerator.flow(x_valid, y=y_valid)
```

c. [LO 3, 15 poin] Buatlah arsitektur baseline sesuai dengan gambar arsitektur AlexNet berikut
ini: (Catatan: Activation function tiap hidden layer menggunakan ReLU)

```python
Baseline_Alexnet = tf.keras.Sequential()

# Convolutional layer 1
Baseline_Alexnet.add(tf.keras.layers.Conv2D(64, (5, 5), strides=(4, 4), activation='relu', input_shape=(64, 64, 3)))
Baseline_Alexnet.add(tf.keras.layers.BatchNormalization())
Baseline_Alexnet.add(tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Convolutional layer 2
Baseline_Alexnet.add(tf.keras.layers.Conv2D(256,( 3, 3), strides=(1, 1), padding='same', activation='relu'))
Baseline_Alexnet.add(tf.keras.layers.BatchNormalization())
Baseline_Alexnet.add(tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Convolutional layer 3
Baseline_Alexnet.add(tf.keras.layers.Conv2D(384, (3, 3), strides=(1, 1), padding='same', activation='relu'))

# Convolutional layer 4
Baseline_Alexnet.add(tf.keras.layers.Conv2D(384, (3, 3), strides=(1, 1), padding='same', activation='relu'))

# Convolutional layer 5
Baseline_Alexnet.add(tf.keras.layers.Conv2D(192, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Baseline_Alexnet.add(tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Flatten layer
Baseline_Alexnet.add(tf.keras.layers.Flatten())

# Fully connected layer 1
Baseline_Alexnet.add(tf.keras.layers.Dense(4096, activation='relu'))
Baseline_Alexnet.add(tf.keras.layers.Dropout(0.5))

# Fully connected layer 2
Baseline_Alexnet.add(tf.keras.layers.Dense(4096, activation='relu'))
Baseline_Alexnet.add(tf.keras.layers.Dropout(0.5))

# Output layer
Baseline_Alexnet.add(tf.keras.layers.Dense(num_classes, activation='softmax'))

```

```python
#compile alexnet
Baseline_Alexnet.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
print(Baseline_Alexnet.summary())
```

```python
#training dan result
history = Baseline_Alexnet.fit(trainGenerator, epochs=10, verbose=1, validation_data=valGenerator)
```

```python
Tune5 = tf.keras.Sequential()

# input dan rescaling
Tune5.add(tf.keras.layers.Rescaling(scale=1. / 255, input_shape=(img_size)))

# Convolutional layer 1
Tune5.add(tf.keras.layers.Conv2D(64, (5, 5), strides=(4, 4), activation='relu', input_shape=(64, 64, 3)))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))
Tune5.add(tf.keras.layers.Dropout(0.3))

# Convolutional layer 2
Tune5.add(tf.keras.layers.Conv2D(256, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Conv2D(256, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
Tune5.add(tf.keras.layers.Dropout(0.4))

# Convolutional layer 3
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=(2, 2)))
Tune5.add(tf.keras.layers.Dropout(0.4))

# Convolutional layer 4
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Conv2D(512, (3, 3), strides=(1, 1), padding='same', activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())

Tune5.add(tf.keras.layers.Dropout(0.4))

# Flatten layer
Tune5.add(tf.keras.layers.Flatten())

# Fully connected layer 1
Tune5.add(tf.keras.layers.Dense(4096, activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Dropout(0.5))

# Fully connected layer 2
Tune5.add(tf.keras.layers.Dense(4096, activation='relu'))
Tune5.add(tf.keras.layers.BatchNormalization())
Tune5.add(tf.keras.layers.Dropout(0.5))

# Output layer
Tune5.add(tf.keras.layers.Dense(num_classes, activation='softmax'))

Tune5.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```
